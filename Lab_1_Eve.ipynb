```python
from sklearn.decomposition import PCA
```

```python
fake_df_reduced = PCA(n_components = 2).fit_transform(fake_df_counts.toarray())
```

```python
fake_df_reduced.shape
```

```python
categories = ['News', 'politics', 'left-news', 'Government News', 'Middle-east', 'US_News']
```

```python
for category in categories:
    print(category, (fake_df['subject'] == category).sum())
```

```python
col = ['coral', 'blue', 'black', 'orange', 'red', 'green']
markers = ['o', 's', '*', 'x', '+', 'd']

# plot
fig = plt.figure(figsize = (25,10))
ax = fig.subplots()

# for c, category in zip(col, categories):
for c, marker, category in zip(col, markers, categories):
    xs = fake_df_reduced[fake_df['subject'] == category].T[0]
    ys = fake_df_reduced[fake_df['subject'] == category].T[1]

    ax.scatter(xs, ys, c = c, marker=marker, label=category)

ax.grid(color='gray', linestyle=':', linewidth=2, alpha=0.2)
ax.set_xlim([fake_df_reduced[:,0].min() - 1, fake_df_reduced[:,0].max() + 1])
ax.set_ylim([fake_df_reduced[:,1].min() - 1, fake_df_reduced[:,1].max() + 1])
ax.set_xlabel('\nX Label')
ax.set_ylabel('\nY Label')

plt.show()
```
